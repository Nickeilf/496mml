```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('pdf', 'svg')
_1a_left = -0.3
_1a_right = 1.3
```

```python
def get_data():
    N = 25
    X = np.reshape(np.linspace(0, 0.9, N), (N, 1))
    Y = np.cos(10*X**2) + 0.1 * np.sin(100*X)
    return X,Y

X, Y = get_data()
```

```python
def plot_points(X, Y, x_left, x_right):
    plt.scatter(X, Y, c='black', marker='.')
    plt.xlim((x_left, x_right))
```

```python
def plot_curve(X, Y, x_left, x_right, label):
    plt.plot(X, Y, label=label)
    plt.xlim((x_left, x_right))
    plt.ylim((-1.5,1.5))
```

# 1.a Polynomial Linear Regression

```python
def init_phi(X, K, N):
    phi = np.zeros((K+1, N))
    phi[0] = 1
    for i in range(K):
        phi[i+1] = (X**(i+1)).reshape(N)
    return phi.T
```

```python
plot_points(X, Y, _1a_left, _1a_right)
for K in [0, 1, 2, 3, 11]:
    # train
    phi = init_phi(X, K, 25)
    tmp = np.dot(phi.T, phi)
    inv_tmp = np.linalg.inv(tmp)
    theta = np.dot(np.dot(inv_tmp, phi.T), Y)

    # predict
    x_plot = np.linspace(-0.3, 1.3, 200)
    phi_plot = init_phi(x_plot, K, 200)
    y_plot = np.dot(theta.T, phi_plot.T).reshape(200)
    plot_curve(x_plot, y_plot, _1a_left, _1a_right, "order = {}".format(K))
plt.title("Polynomial Linear Regression")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# 1.b Linear Regression with Trigonometric basis function

```python
def init_tri(X, K, N):
    tri = np.zeros((N, 2*K+1))
    tri[:,0] = 1
    for i in range(1,K+1):
        tri[:, 2*i] = np.cos(2*np.pi*i*X[:, 0])
        tri[:, 2*i-1] = np.sin(2*np.pi*i*X[:, 0])
    return tri
```

```python
N=25
plt.figure(figsize=(10,7))
for K in [1, 11]:
    # train
    tri = init_tri(X, K, 25)
    tmp = np.dot(tri.T, tri)
    inv_tmp = np.linalg.inv(tmp)
    theta = np.dot(np.dot(inv_tmp, tri.T), Y)

    # predict
    x_plot = np.linspace(-1, 1.2, 200).reshape(200,1)
    
    tri_plot = init_tri(x_plot, K, 200)
    y_plot = np.dot(theta.T, tri_plot.T).reshape(200)
    plot_curve(x_plot, y_plot, _1a_left, _1a_right, "order = {}".format(K))
plt.title("Linear Regression with trigonometric basis functions")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('pic/1.b.png')
```

# 1.c LOO CV 

```python
def plot_error(sigma, error, X):
    plt.plot(X, error, label='average squared test error')
    plt.plot(X, sigma, label='maximum likelihood sigma^2')
    plt.ylim((0,0.6))
```

```python
N = 25
plt.figure(figsize=(10,7))
test_errors = np.zeros(11)
train_errors = np.zeros(11)
for K in range(11):
    test_error = 0
    train_error = 0
    for i in range(N):
        X_train = np.delete(X, i, 0)
        y_train = np.delete(Y, i, 0)
        X_test = X[i].reshape((1,1))
        y_test = Y[i].reshape((1,1))
        
        tri = init_tri(X_train, K, N-1)
        tmp = np.dot(tri.T, tri)
        inv_tmp = np.linalg.inv(tmp)
        theta = np.dot(np.dot(inv_tmp, tri.T), y_train)
        y_predict_train = np.dot(theta.T, tri.T).reshape(N-1,1)
        
        #test
        tri_test = init_tri(X_test, K, 1)
        y_predict = np.dot(theta.T, tri_test.T).reshape(1,1)
        
        test_error += (y_predict - y_test)**2
        y_difference = y_predict_train - y_train
        train_error += np.dot(y_difference.T, y_difference)
        
    test_errors[K] = test_error[0][0]/N
    train_errors[K] = train_error[0][0]/(N-1)/N

plot_error(train_errors, test_errors, np.arange(11)+1)
plt.title('overfitting with LOO')
plt.legend()
plt.xlabel('Order')
plt.savefig('pic/1.c.png')
```

# 2.Ridge Regression

```python
def init_gaussian(X, K, N, means):
    gaussian = np.zeros((N, K+1))
    gaussian[:,0]=1
    for i in range(1,K+1):
        gaussian[:,i] = np.exp(-(X-means[i-1])**2/2/0.1/0.1)[:,0]
    return gaussian
```

```python
N=25
K=20
plt.figure(figsize=(10,7))
plot_points(X, Y, _1a_left, _1a_right)
means = np.linspace(0, 1, K)
_lambdas = [0.00001, 0.1, 10000]
for _lambda in _lambdas:
    # train
    gau = init_gaussian(X, K, N, means)
    tmp = np.dot(gau.T, gau) + np.eye(K+1)*_lambda
    inv_tmp = np.linalg.inv(tmp)
    W = np.dot(np.dot(inv_tmp, gau.T), Y)
    
    x_plot = np.linspace(-0.3, 1.3, 200).reshape(200,1)
    gau_plot = init_gaussian(x_plot, K, 200, means)
    y_plot = (np.dot(W.T, gau_plot.T)+_lambda*np.dot(W.T,W).reshape(1)).reshape(200,1)
    plot_curve(x_plot, y_plot,_1a_left,_1a_right, label="lambda = {}".format(_lambda))

plt.title("Ridge Regression")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('pic/2.b.png')
```
